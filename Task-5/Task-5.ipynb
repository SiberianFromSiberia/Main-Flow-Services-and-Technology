```python
questions=["1.How many people have heart disease and how many people doesnt have heart disease?",
          "2.People of which sex has most heart disease?",
          "3.People of which sex has which has type of chest pain most?",
          "4.People with chest pain are most prone to have a heart disease?"]
questions
```

```python
#Answer for 1st qn
df.target.value_counts()
```

```python
df.target.value_counts().plot(kind='bar',color=['orchid','salmon'])
plt.title("Heart Disease Values")
plt.xlabel("1=Heart Disease,0=No Heart Disease")
plt.ylabel("Amount");
```

```python
df.target.value_counts().plot(kind='pie',figsize=(8,6))
plt.legend(['Disease','No Disease']);
```

```python
#'0' represents 'Female'
#'1' represents 'male'
#'0' represents 'No Disease'
#'1' represents 'Disease'
#checking how many male and female are in dataset
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind='pie',figsize=(8,6))
plt.title('Male Female Ratio')
plt.legend(['Male','Female']);
```

```python
#Answer for 2ND Qn
pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x='target',data=df,hue='sex')
plt.title('Heart Disease frequency for sex')
plt.xlabel('0=No Heart Disease,1=Heart Disease');
```

```python
#Answer for 3rd Qn
df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind='bar',color=['salmon','lightskyblue','springgreen','khaki'])
plt.title("Chest Pain v/s Count");
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind='bar',color=['coral','pink','lightblue','plum','palegreen'])
plt.title('Type of Chest Pain for Sex')
plt.xlabel('0=Female,1=Male');
```

```python
#Answer For 4th qn
pd.crosstab(df.cp,df.target)
sns.countplot(x='cp',data=df,hue='target');
```

```python
#taking look at our age column
#create a distribution plot with normal distribution curve
sns.displot(x='age',data=df,bins=30,kde=True);
```

```python
#58-59 years old are most in the dataset
#plotting another distribution plot for maximum heart rate
sns.displot(x='thalach',data=df,bins=30,kde=True,color='green');
```
