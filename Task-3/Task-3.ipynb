```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df=pd.read_csv("householdtask3.csv")
```

```python
df.head(10)
```

```python
#Scatter plot
plt.scatter(df['year'],df['own'])
plt.title("Scatter plot")
plt.xlabel("year")
plt.ylabel("own")
plt.show()
```

```python
#Line chart
plt.plot(df['year'])
plt.plot(df['own'])
plt.title("Line Chart")
plt.xlabel("year")
plt.ylabel("own")
plt.show()
```

```python
#Bar Graph
plt.bar(df['year'],df['own'])
plt.title("Bar Graph")
plt.xlabel("year")
plt.ylabel("own")
plt.show()
```

```python
#Histogram
plt.hist(df['income'])
plt.title("Histogram")
plt.show()
```
